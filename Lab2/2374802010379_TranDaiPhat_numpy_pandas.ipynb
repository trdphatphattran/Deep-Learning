### Bài 1: 
Cho ma trận = [[99, 99, 99]

               [99, 99, 99]
               
               [99, 99, 99]]

Hãy tưởng tượng ô 99 trống

Giả sử 1 là x, 0 là o

Lấy đầu vào từ phía x và phía o luân phiên.

Nếu đầu vào từ phía x là (0, 0) thì ma trận sẽ trở thành 

               [[x, 99, 99]

               [99, 99, 99]
               
               [99, 99, 99]]

Nếu đầu vào từ phía o cũng là (0, 0) 

Thông báo họ nhập lại, nếu không thì điền vào ma trận.

Thử thách: Nếu người đầu tiên có 3 phần tử liên tiếp thì dừng chơi.

```python
import numpy as np

# Khởi tạo bàn cờ
board = np.full((3, 3), 99)

def print_board():
    print(board)
    print()

def check_win(player):
    # Kiểm tra hàng
    if np.any(np.all(board == player, axis=1)):
        return True

    # Kiểm tra cột
    if np.any(np.all(board == player, axis=0)):
        return True

    # Kiểm tra đường chéo
    if np.all(np.diag(board) == player):
        return True
    if np.all(np.diag(np.fliplr(board)) == player):
        return True

    return False


current_player = 1  

while True:
    print_board()
    symbol = "x" if current_player == 1 else "o"
    print(f"Lượt của {symbol}")

    row = int(input("Nhập row (0-2): "))
    col = int(input("Nhập col (0-2): "))

    # Kiểm tra ô trống
    if board[row, col] != 99:
        print("Ô đã được chọn, vui lòng nhập lại!\n")
        continue

    # Điền vào ma trận
    board[row, col] = current_player

    # Kiểm tra thắng
    if check_win(current_player):
        print_board()
        print(f"{symbol.upper()} chiến thắng")
        break

    # Đổi lượt
    current_player = 0 if current_player == 1 else 1
```

### Bài 4:
Tạo một mảng numpy có kích thước 150 x 5. Giả sử mẫu này chứa 150 mẫu dữ liệu của sinh viên bao gồm chiều cao, cân nặng, tuổi, lương và điểm trung bình.

Tách mảng dựa trên 4 cột đầu tiên thành một biến có tên là x và cột cuối cùng thành y.

Tách mảng x thành x_train và x_test, trong đó x_train chứa 70% dữ liệu và tách y thành y_train và y_test, trong đó y_train chứa 70% dữ liệu.

Tạo 10 tập dữ liệu x_train không chồng chéo nhau

```python
import numpy as np

# 1. Tạo dữ liệu 150 x 5
np.random.seed(42)
data = np.random.rand(150, 5)

# 2. Tách x và y
x = data[:, :4] # 4 cột đầu
y = data[:, 4] # cột cuối

# 3. Chia train / test (70% - 30%)
n_samples = x.shape[0]
train_size = int(0.7 * n_samples)

indices = np.random.permutation(n_samples)

train_idx = indices[:train_size]
test_idx  = indices[train_size:]

x_train = x[train_idx]
x_test  = x[test_idx]

y_train = y[train_idx]
y_test  = y[test_idx]

# 4. Tạo 10 tập x_train không chồng chéo
x_train_splits = np.array_split(x_train, 10)

print("x shape:", x.shape)
print("y shape:", y.shape)

print("x_train shape:", x_train.shape)
print("x_test shape:", x_test.shape)

for i, part in enumerate(x_train_splits):
    print(f"x_train_split {i+1}: {part.shape}")
```

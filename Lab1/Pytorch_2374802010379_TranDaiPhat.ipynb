## Bài 1
Cho y = 5x^5 + 6x^3 - 3x + 1

Cho biết độ dốc của đa thức trên ở điểm nào


```python
import torch

x = torch.tensor(1.0, requires_grad=True)

y = 5*x**5 + 6*x**3 - 3*x + 1

y.backward()

print(f"Hàm số: y = 5x^5 + 6x^3 - 3x + 1")
print(f"Độ dốc của hàm số tại x = {x.item()} là: {x.grad.item()}")
```

## Bài 2
Tạo một tensor ban đầu có giá trị là 2

Định nghĩa hàm số và tính gradient y = x^3 + 2x^2 + 5x + 1

Hãy tính dy/dx tại giá trị của x?

Dùng phương pháp Gradient Descent với Learning Rate alpha = 0.1 để cập nhập giá trị x trong 10 vòng lặp

```python
import torch

x = torch.tensor(2.0, requires_grad=True)
alpha = 0.1
epochs = 10

print(f"Giá trị ban đầu: x = {x.item():.4f}")
print("-" * 30)

for epoch in range(epochs):
    y = x**3 + 2*x**2 + 5*x + 1
    
    if x.grad is not None:
        x.grad.zero_()
        
    y.backward()
    
    current_grad = x.grad.item()
    
    with torch.no_grad():
        x -= alpha * x.grad
        
    print(f"Vòng lặp {epoch+1}: dy/dx = {current_grad:>7.2f} | x mới = {x.item():.4f}")

print("-" * 30)
print(f"Giá trị x cuối cùng sau 10 vòng lặp: {x.item():.4f}")
```

## Bài 3
Tạo một tập dữ liệu giả lập với x là số giờ học ngẫu nhiên từ 1 đến 10 và y là số điểm được tính theo công thức: y = 3x + 5 + noise (với noise là một giá trị ngẫu nhiên nhỏ).

1. Khởi tạo tham số w và b ngẫu nhiên với requires_grad = True.
2. Tính MSE.
3. Tính gradient.
4. Cập nhật tham số w và b bằng Gradient Descent với Learning Rate alpha = 0.01
5. Lặp lại quá trình trên trong 100 vòng lặp để xem sự hội tụ của mô hình.

```python
import torch

torch.manual_seed(42)

x = torch.linspace(1, 10, 100).reshape(-1, 1)

noise = torch.randn(x.size()) * 1.5
y = 3 * x + 5 + noise

w = torch.randn(1, requires_grad=True)
b = torch.randn(1, requires_grad=True)

alpha = 0.01  
epochs = 100  

print(f"Giá trị khởi tạo: w = {w.item():.4f}, b = {b.item():.4f}")
print("-" * 50)

for epoch in range(epochs):
    y_pred = x * w + b

    mse_loss = torch.mean((y_pred - y)**2)
    
    mse_loss.backward()
    
    with torch.no_grad():
        w -= alpha * w.grad
        b -= alpha * b.grad
        
        w.grad.zero_()
        b.grad.zero_()
    
    if (epoch + 1) % 10 == 0:
        print(f"Vòng lặp {epoch+1:3d} | MSE: {mse_loss.item():.4f} | w: {w.item():.4f} | b: {b.item():.4f}")

print("-" * 50)
print(f"KẾT QUẢ CUỐI CÙNG:")
print(f"MSE cuối cùng: {mse_loss.item():.4f}")
print(f"Công thức học được: y = {w.item():.4f}x + {b.item():.4f}")
```
